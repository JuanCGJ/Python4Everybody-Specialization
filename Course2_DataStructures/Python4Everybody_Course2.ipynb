## 📌 Week 1 — Strings & String Methods

### Assignment Description
Extract a floating-point number embedded inside a string of email metadata.

### Concepts Covered
- **`str.find()`** — locate the position of a character inside a string
- **String slicing `[i:]`** — extract a substring from a given position to the end
- **`float()`** — convert a string to a decimal number

### Key Takeaway
> Strings are sequences — you can navigate them using index positions. `find()` returns the index of the first match, which we then use to slice out the data we need.

````python
# Week 1 — Extract number from string
text = "X-DSPAM-Confidence:    0.8475"
ipos = text.find('0')
num = text[ipos:]
print(float(num))
````

**Output:** `0.8475`

## 📌 Week 3 (Assignment 1) — Reading Files & String Methods

### Assignment Description
Open `words.txt`, read it line by line, strip trailing whitespace, and print each line in uppercase.

### Concepts Covered
- **`open()`** — opening a file for reading
- **`for line in file`** — iterating over lines in a file
- **`rstrip()`** — removing trailing whitespace/newlines
- **`upper()`** — converting a string to uppercase

### Key Takeaway
> Python treats files as sequences of lines — we can iterate over them directly with a `for` loop without loading everything into memory at once.

> 📁 **Required file:** `words.txt` (included in this course folder)

````python
# Week 3 (1) — Read file and print lines in uppercase
fname = input("Enter file name: ")  # use: words.txt
fh = open(fname)

for lx in fh:
    ly = lx.rstrip()
    print(ly.upper())
````

**Output:** Each word from `words.txt` printed in uppercase.

## 📌 Week 3 (Assignment 2) — File Parsing & Computing Averages

### Assignment Description
Parse `mbox-short.txt` and compute the average spam confidence score from lines starting with `X-DSPAM-Confidence:`.

### Concepts Covered
- **`startswith()`** — filtering lines by prefix
- **`continue`** — skipping irrelevant lines
- **String slicing** — extracting numeric values from fixed-format text
- **Accumulator pattern** — summing values and counting iterations

### Key Takeaway
> The accumulator pattern (`total += value`, `count += 1`) is the foundation of computing sums and averages from large datasets.

> 📁 **Required file:** `mbox-short.txt` (included in this course folder)

````python
# Week 3 (2) — Average spam confidence from file
fname = input("Enter file name: ")  # use: mbox-short.txt
fh = open(fname)
count = float(0)
add = float(0)

for line in fh:
    if not line.startswith("X-DSPAM-Confidence:"):
        continue
    num = line[20:]
    fnum = float(num)
    count = count + 1
    add = add + fnum

print('Average spam confidence:', add / count)
````

**Output:** `Average spam confidence: 0.7507185185185187`

## 📌 Week 4 (Assignment 1) — Lists & Deduplication

### Assignment Description
Read `romeo.txt`, extract every unique word, sort them alphabetically, and print the resulting list.

### Concepts Covered
- **`split()`** — breaking a line into a list of words
- **`append()`** — adding items to a list
- **Deduplication** — checking `if w in lst` before appending
- **`sort()`** — sorting a list in place

### Key Takeaway
> Lists preserve order and allow duplicates by default. Checking membership with `in` before appending is a simple but powerful deduplication pattern.

> 📁 **Required file:** `romeo.txt` (included in this course folder)

````python
# Week 4 (1) — Unique sorted word list from file
fname = input("Enter file name: ")  # use: romeo.txt
fh = open(fname)
lst = list()

for line in fh:
    line = line.rstrip()
    line = line.split()
    for w in line:
        if w in lst:
            continue
        lst.append(w)

lst.sort()
print(lst)
````

**Output:** `['Arise', 'But', 'It', 'Juliet', 'Who', 'already', ...]`

## 📌 Week 4 (Assignment 2) — Lists & Line Filtering

### Assignment Description
Read `mbox-short.txt`, filter lines starting with `From:`, extract each sender's email address, print it, and count the total.

### Concepts Covered
- **`startswith()`** — filtering lines
- **`split()`** — splitting a line into tokens
- **Index access `x[1]`** — getting the email address (second token)
- **Counter pattern** — incrementing a variable to count matches

### Key Takeaway
> Combining `split()` with index access is a powerful way to extract structured fields from text data — a pattern used constantly in log parsing and data engineering.

> 📁 **Required file:** `mbox-short.txt` (included in this course folder)

````python
# Week 4 (2) — Extract email addresses from From: lines
fname = input("Enter file name: ")  # use: mbox-short.txt
fh = open(fname)
count = 0

for line in fh:
    line.rstrip()
    if not line.startswith('From:'):
        continue
    x = line.split()
    print(x[1])
    count = count + 1

print("There were", count, "lines in the file with From as the first word")
````

**Output:** Email addresses followed by total count.

## 📌 Week 5 — Dictionaries & Finding the Most Frequent Item

### Assignment Description
Read `mbox-short.txt`, count how many emails each sender sent, then find and print the most frequent sender.

### Concepts Covered
- **`dict()`** — creating a dictionary
- **`.get(key, default)`** — safely incrementing a count without KeyError
- **Iterating `.items()`** — looping through key-value pairs
- **Finding the maximum** — tracking the largest value seen

### Key Takeaway
> The pattern `d[key] = d.get(key, 0) + 1` is one of the most useful idioms in Python — it's the foundation of frequency counting and histogram building.

> 📁 **Required file:** `mbox-short.txt` (included in this course folder)

````python
# Week 5 — Most frequent email sender using a dictionary
name = input("Enter file:")  # use: mbox-short.txt
if len(name) < 1: name = "mbox-short.txt"
handle = open(name)
mails = dict()

for line in handle:
    if not line.startswith('From:'):
        continue
    wds = line.split()
    for l in wds:
        if not l == 'From:':
            mails[l] = mails.get(l, 0) + 1

largo = -1
correo = None

for k, v in mails.items():
    if v > largo:
        largo = v
        correo = k

print(correo, largo)
````

**Output:** `zqian@umich.edu 195`

## 📌 Week 6 — Tuples & Sorted Output

### Assignment Description
Read `mbox-short.txt`, extract the hour from each timestamp, count emails per hour, and print results sorted by hour.

### Concepts Covered
- **Tuples** — immutable pairs used for sorting `(key, value)`
- **`sorted()`** — sorting a list of tuples by first element
- **String slicing** — extracting the hour `h[0:2]` from a timestamp
- **Dictionary + list** — building a frequency table then sorting it

### Key Takeaway
> Converting a dictionary to a list of `(key, value)` tuples and calling `sorted()` is the classic Python pattern for producing ordered output from unordered data.

> 📁 **Required file:** `mbox-short.txt` (included in this course folder)

````python
# Week 6 — Email count per hour using tuples and sorted output
name = input("Enter file:")  # use: mbox-short.txt
if len(name) < 1: name = "mbox-short.txt"
handle = open(name)
diction = dict()
hour = list()

for line in handle:
    if not line.startswith('From '):
        continue
    line = line.split()
    h = line[5]
    hour.append(h)

for c in hour:
    num = c[0:2]
    diction[num] = diction.get(num, 0) + 1

final = list()
for k, v in diction.items():
    fin = (k, v)
    final.append(fin)
    orden = sorted(final)

for k, v in orden:
    print(k, v)
````

**Output:**
```
04 3
06 1
10 3
11 6
...
```
