## 📌 Week 5 — Conditional Statements (if / elif / else)

### Assignment Description
Write a grade classifier that reads a score between `0.0` and `1.0` and prints the corresponding letter grade. If the score is out of range, print `"Out of range"`.

### Grading Scale
| Score | Grade |
|-------|-------|
| >= 0.9 | A |
| >= 0.8 | B |
| >= 0.7 | C |
| >= 0.6 | D |
| < 0.6  | F |

### Concepts Covered
- **`if / elif / else`** — branching logic
- **Nested conditionals** — checking ranges with multiple conditions
- **Comparison operators** — `>=`, `<`, `<=`

### Key Takeaway
> Python evaluates `if/elif` blocks **top to bottom** and stops at the first `True` condition. Order matters!

````python
# Week 5 — Grade Classifier
sc = input("Enter Score: ")
score = float(sc)

if score >= 0.0:
    if score <= 1.0:
        if score >= 0.9:
            print("A")
        elif score >= 0.8:
            print("B")
        elif score >= 0.7:
            print("C")
        elif score >= 0.6:
            print("D")
        elif score < 0.6:
            print("F")
    else:
        print("Out of range")

if score < 0.0:
    print("Out of range")
````

````python
# ✅ Expected Output — Week 5
# Simulating different score inputs
def grade(score):
    if score >= 0.0:
        if score <= 1.0:
            if score >= 0.9: return 'A'
            elif score >= 0.8: return 'B'
            elif score >= 0.7: return 'C'
            elif score >= 0.6: return 'D'
            elif score < 0.6: return 'F'
        else: return 'Out of range'
    if score < 0.0: return 'Out of range'

for score in [0.95, 0.85, 0.73, 0.61, 0.45, 1.5, -0.1]:
    print(f'Score {score} → {grade(score)}')
````

**Example runs:**
```
Enter Score: 0.95  →  A
Enter Score: 0.73  →  C
Enter Score: 1.5   →  Out of range
```

## 📌 Week 6 — Functions & Business Logic

### Assignment Description
Rewrite the pay calculator using a **function** `computepay(h, r)`. This time, apply **overtime pay**: any hours worked beyond 40 are paid at **1.5× the regular rate**.

### Overtime Rule
```
If hours <= 40:  pay = hours × rate
If hours >  40:  pay = (40 × rate) + (extra_hours × rate × 1.5)
```

### Concepts Covered
- **`def`** — defining reusable functions
- **Parameters & return values** — passing data in and out of functions
- **Business logic** — encoding real-world rules into code

### Key Takeaway
> Functions let us **reuse logic** without repeating code. This is the foundation of clean, maintainable programs.

````python
# Week 6 — Pay Calculator with Overtime
def computepay(h, r):
    b = 40.0    # base hours threshold
    a = 1.5     # overtime multiplier
    if h <= b:
        p = h * r
    else:
        p = b * r
        add = (h - 40) * r * a
        p = float(p + add)
    return p

hrs = input("Enter Hours:")
h = float(hrs)
hr = input("Enter Hourly Rate: ")
r = float(hr)

p = computepay(h, r)
print("Pay", p)
````

````python
# ✅ Expected Output — Week 6
# Simulating different hour/rate inputs
def computepay(h, r):
    b, a = 40.0, 1.5
    if h <= b:
        p = h * r
    else:
        p = b * r + (h - 40) * r * a
    return float(p)

cases = [(40, 10), (45, 10), (50, 20)]
for h, r in cases:
    print(f'Hours: {h} | Rate: {r} → Pay: {computepay(h, r)}')
````

**Example runs:**
```
Enter Hours: 40  |  Enter Rate: 10  →  Pay 400.0  (no overtime)
Enter Hours: 45  |  Enter Rate: 10  →  Pay 475.0  (5 overtime hours × 1.5)
```

## 📌 Week 7 — Loops & Error Handling

### Assignment Description
Write a program that repeatedly asks the user to enter numbers. When the user types `"done"`, the program stops and prints the **maximum** and **minimum** values entered. Handle invalid inputs gracefully.

### Concepts Covered
- **`while True`** — infinite loop controlled by a `break`
- **`break`** — exiting a loop on a condition
- **`try / except`** — catching errors without crashing
- **`None`** — representing "no value yet" as an initial state
- **`continue`** — skipping the rest of the loop iteration

### Key Takeaway
> `try/except` makes programs **robust**: instead of crashing on bad input, we catch the error and ask the user to try again.

````python
# Week 7 — Max & Min Finder with Error Handling
largest = None
smallest = None

while True:
    num = input("Enter a number: ")
    if num == "done":
        break
    try:
        inum = int(num)
    except:
        print("Invalid input")
        continue

    if largest is None:
        largest = inum
    elif inum > largest:
        largest = inum

    if smallest is None:
        smallest = inum
    elif inum < smallest:
        smallest = inum

print("Maximum is", largest)
print("Minimum is", smallest)
````

````python
# ✅ Expected Output — Week 7
# Simulating input sequence: 5, 3, 'abc', 9, 'done'
inputs = ['5', '3', 'abc', '9', 'done']
largest = None
smallest = None

for num in inputs:
    print(f'> {num}')
    if num == 'done':
        break
    try:
        inum = int(num)
    except:
        print('Invalid input')
        continue
    if largest is None or inum > largest: largest = inum
    if smallest is None or inum < smallest: smallest = inum

print('Maximum is', largest)
print('Minimum is', smallest)
````

**Example run:**
```
Enter a number: 5
Enter a number: 3
Enter a number: abc  →  Invalid input
Enter a number: 9
Enter a number: done
Maximum is 9
Minimum is 3
```
